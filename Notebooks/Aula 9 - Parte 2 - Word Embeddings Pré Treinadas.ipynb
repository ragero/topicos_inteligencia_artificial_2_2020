# Carregando as Word Embeddings

```
modelo_wv = KeyedVectors.load_word2vec_format('cbow_s300.txt')
```

# Consultando o Vetor de Uma Palavra

```
modelo_wv.get_vector('tecnologia')
```

# Obtendo as palavra mais similares

```
modelo_wv.most_similar('automóvel')
```

```
modelo_wv.most_similar('tecnologia')
```

#Somando vetores e consultando as palavras mais similares

```
modelo_wv.most_similar(positive=['tecnologia','telefone'])
```
