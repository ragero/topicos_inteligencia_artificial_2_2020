```
#Nesta aula iremos utilizar o OpenCV 
import cv2
#Importando a Bilioteca MatplotLIB para plotar as imagens
import matplotlib.pyplot as plt
```

# Lendo uma Imagem

```
imagem = cv2.imread('/content/drive/MyDrive/Imagens/homer.jpg')
```

```
#Por padrão, as imagens no openCV são lidas no padrão BGR
plt.imshow(imagem)
```

```
#Covertendo uma imagem para RGB
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```
plt.imshow(imagem_rgb)
```

```
#Exibindo o "valor da imagem" (estrutura de armazenamento)
imagem_rgb
```

```
#Tensor (altura, largura, caracterísicas de cada pixel - RGB): 
imagem_rgb.shape
```

# Convertendo para escala de cinza

```
imagem_gray = cv2.cvtColor(imagem_rgb, cv2.COLOR_RGB2GRAY)
```

```
imagem_gray
```

```
#Matriz (altura e largura)
imagem_gray.shape
```

```
plt.imshow(imagem_gray, cmap='gray')
```

```
#Pode-se fazer a leitura direto em escala de cinza
imagem_gray = cv2.imread('/content/drive/MyDrive/Imagens/homer.jpg', 0)
```

```
plt.imshow(imagem_gray, cmap='gray')
```

# Convertendo para Branca e Preta

```
#Para converter a imagem, passa-se: 
#- A imagem
#- O limiar de corte (valores abaixo desse limiar serão mapeados para 0)
#- O valor atribuído aos pixels que excederem o limiar
#- O método de limiarização
ret,imagem_binaria = cv2.threshold(imagem_gray,127,1,cv2.THRESH_BINARY)
```

```
imagem_binaria
```

```
plt.imshow(imagem_binaria, cmap='gray')
```

# Filtros nas Imagens

```
imagem_eq = cv2.equalizeHist(imagem_gray)
plt.imshow(imagem_eq, cmap='gray')
```

```
imagem_canny = cv2.Canny(imagem_gray, 100,220)
plt.imshow(imagem_canny, cmap='gray')
```

#Redimensionando as Imagens

```
imagem_resized = cv2.resize(imagem_gray, (1000, 1000), interpolation = cv2.INTER_AREA)
```

```
plt.imshow(imagem_resized, cmap='gray')
```

#Salvando a imagem

```
cv2.imwrite('/content/drive/MyDrive/Imagens/homer_processsed.jpg', imagem_resized)
```
