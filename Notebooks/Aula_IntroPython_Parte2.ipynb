# Orientação a Objetos

```
class Pessoa(object): 

  def __init__(self, nome, cpf, idade): 
    self.nome = nome
    self.cpf = cpf
    self.idade = idade
  
  def imprime_infos(self): 
    print('Nome:', self.nome)
    print('CPF:', self.cpf)
    print('Idade:', self.idade)

  def __eq__(self, outro): 
    if(self.nome == outro.nome) and (self.cpf == outro.cpf) and (self.idade == outro.idade):
      return True
    else:
      return False
  
  def __lt__(self, outro):
    if self.idade < outro.idade:
      return True
    else:
      return False
  
  def __str__(self): 
    return f'{self.nome},{self.cpf},{self.idade}'
  
  def __repr__(self):
    return f'{self.nome},{self.cpf},{self.idade}'
```

```
class Aluno(Pessoa): 

  def __init__(self, nome, cpf, idade, curso, rga):
    super().__init__(nome,cpf,idade)
    self.curso = curso
    self.rga = rga

  def imprime_infos(self):
    super().imprime_infos()
    print('Curso:', self.curso)
    print('RGA:', self.rga)
    
```

```
class Impressora(object):

  descricao = 'Essa é uma classe para impressoras'

  def __init__(self, marca, qtd_folhas): 
    self.__marca = marca
    self.__qtd_folhas = qtd_folhas

  @property
  def qtd_folhas(self):
    return f'A impressora {self.__marca} possui {self.__qtd_folhas} folhas'

  @qtd_folhas.setter
  def qtd_folhas(self,valor): 
    self.__qtd_folhas = valor

  @staticmethod
  def imprime_descricao():
    print(f'Descrição da classe: {Impressora.descricao}')

  def imprime_infos(self): 
    print(f'Marca: {self.__marca}')
    print(f'Qtd de folhas: {self.__qtd_folhas}')
```

```
hp = Impressora('HP', 100)
```

```
hp.qtd_folhas
```

```
hp.qtd_folhas = 200
```

```
hp.qtd_folhas
```

```
Impressora.descricao
```

```
Impressora.imprime_descricao()
```
